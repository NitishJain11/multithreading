```python
#Q1-> what is multithrading in python ? Why is it used ? Name the module used to handle threads in python. 

# Answer->

# Multithreading in Python refers to the ability of a program to simultaneously execute multiple threads of 
# execution within the same process. A thread is a lightweight sub-process that runs independently of the main
# thread, and can perform tasks in parallel with other threads.

#Multithreading is useful for improving the performance and responsiveness of a program by allowing it to perform 
# multiple tasks in parallel. For example, a program that needs to perform a long-running computation could create a
# separate thread to perform the computation in the background, while the main thread continues to respond to user 
# input and update the user interface.

# #The module used to handle threads in Python is called threading. It provides a simple and powerful interface for
# creating and managing threads within a Python program. The threading module allows developers to create new threads,
# start them, wait for them to finish, and communicate between threads using various synchronization primitives such as
# locks, semaphores, and events. The threading module also includes other useful features such as timers and daemon threads.
# Overall, the threading module is an essential tool for developing concurrent and parallel programs in Python.

import threading
import logging
logging.basicConfig(filename="softwear01.log",filemode='w',level=logging.DEBUG,format= '%(asctime)s %(threadName)s: %(message)s')
def statment(n):
    for i in range(n):
        logging.debug(f" this is the statemt {i}")
ther = threading.Thread(target = statment(10))
ther.start()
ther.join()
logging.info(ther)
```

```python
# Q4->Write the python program to create two threads . Theads one must print the list of squares and thread
# two must print the list of cubes.

# Answer->

import logging
import threading

logging.basicConfig(filename ="multithrading.log",filemode = 'w',level=logging.DEBUG, format='%(asctime)s %(threadName)s: %(message)s ')

def square_list(n):
    for i in range(n):
        square = [ i**2 for i in range(n)]
    logging.debug(f" This is the Square list {square}")
      

def cube_list(n):
    for i in range(n):
        cube = [i**3 for i in range(10)]
    logging.debug(f" This is the cube list{cube}")
        

if __name__ == '__main__':
 
    thread1 = threading.Thread(target=square_list(10))
  
    thread2 = threading.Thread(target=cube_list(10))

    thread1.start()
    thread2.start()
    thread1.join()
    thread2.join()
```

```python
# Q5-> State advantages and disadvantages of multithreading.


# Answer->

# Multithreading is the ability to execute multiple threads of execution concurrently within a single process. 
# Here are some advantages and disadvantages of multithreading:

# Advantages:

# Improved performance: Multithreading can improve the performance of a program by allowing it to utilize multiple 
# cores of a CPU or multiple CPUs. This can result in faster execution times for CPU-bound tasks.

# Better resource utilization: Multithreading allows a program to make better use of system resources, such as memory 
# and I/O devices. This is because multiple threads can share resources instead of each thread requiring its own set of resources.

# Enhanced responsiveness: Multithreading can improve the responsiveness of a program by allowing it to continue executing other 
# tasks while waiting for I/O operations or other blocking calls to complete.

# Simplified program design: Multithreading can simplify program design by allowing a program to be broken down into smaller,
# more manageable tasks that can be executed concurrently.

# Disadvantages:

# Increased complexity: Multithreading can add complexity to a program, making it more difficult to design, debug, and maintain.
# Multithreading requires careful synchronization and coordination of threads to avoid race conditions and other synchronization issues.

# Increased memory usage: Multithreading can increase the memory usage of a program, as each thread requires its own stack and other resources.

# Increased overhead: Multithreading can introduce additional overhead due to the need for thread creation, synchronization, and context switching.

# Difficult to debug: Multithreaded programs can be difficult to debug, as thread interactions can be unpredictable and difficult to reproduce.

# Potential for deadlock: Multithreaded programs can be prone to deadlock, where multiple threads are waiting for resources that are held by other
# threads, resulting in a deadlock state.
```

```python
# Q-6 -> Explain deadlocks and race conditions.

# Answer->

# Deadlocks and race conditions are two common problems that can occur in multithreaded programs.

# A deadlock occurs when two or more threads are blocked and waiting for each other to release resources that they hold. This can result in a situation where none of the threads can make progress, and the program becomes stuck.


import threading
import logging 
logging.basicConfig(filename="example.log", filemode='w',level =logging.DEBUG,format='%(asctime)s %(threadName)s: %(message)s')

# Two resources to be locked
resource1 = threading.Lock()
resource2 = threading.Lock()

# Thread 1 tries to acquire resource 1, then resource 2
def thread1():
    resource1.acquire()
    resource2.acquire()
    logging.debug("Thread 1: Acquired resources 1 and 2")
    resource2.release()
    resource1.release()

# Thread 2 tries to acquire resource 2, then resource 1
def thread2():
    resource2.acquire()
    resource1.acquire()
    logging.debug("Thread 2: Acquired resources 2 and 1")
    resource1.release()
    resource2.release()

t1 = threading.Thread(target=thread1)
t2 = threading.Thread(target=thread2)
t1.start()
t2.start()
```

```python
# A race condition occurs when two or more threads access a shared resource in a way that produces unexpected results.
# The outcome of the program depends on the order in which the threads execute, and the result can be unpredictable and 
# inconsistent.


import threading
import logging

logging.basicConfig(filename = "example.log",level=logging.DEBUG, format='%(asctime)s %(message)s')

counter = 0
lock = threading.Lock()

def increment():
    global counter
    for i in range(10):
        with lock:
            counter += 1
            logging.debug('Counter is %d', counter)

# Start two threads that increment the counter
t1 = threading.Thread(target=increment, name='Thread 1')
t2 = threading.Thread(target=increment, name='Thread 2')
t1.start()
t2.start()
t1.join()
t2.join()
logging.debug('Final counter value is %d', counter)
```
